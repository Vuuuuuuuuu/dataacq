## Q.12. For the windspeed column, calculate the mean, median, and standard deviation.

```python
windspeed_mean = df['windspeed'].mean()
windspeed_median = df['windspeed'].median()
windspeed_std = df['windspeed'].std()
print(f"Mean of windspeed: {windspeed_mean}")
print(f"Median of windspeed: {windspeed_median}")
print(f"Standard deviation of windspeed: {windspeed_std}")
```

## Q13. Identify any potential outliers in a numerical column of your choice. Explain your approach.


```python
windspeed = df['windspeed']
Q1 = windspeed.quantile(0.25)
Q3 = windspeed.quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = windspeed[(windspeed < lower_bound) | (windspeed > upper_bound)]
print("Outliers in windspeed:")
print(outliers)
```

## Q.14 Find the correlation between numerical columns and discuss any interesting relationships.

```python
numerical_columns = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numerical_columns.corr()
print("Correlation Matrix:")
print(correlation_matrix)
import seaborn as sns
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5, cbar=True)
plt.title("Correlation Matrix of Numerical Columns")
plt.show()
```

## Q.16 In which season (Spring, Summer, Fall, Winter) people rented bikes the most?

```python
file_path = "hour.csv"
df = pd.read_csv(file_path)
season_mapping = {1: "Winter", 2: "Spring", 3: "Summer", 4: "Fall"}
season_rentals = df.groupby("season")["cnt"].sum()
most_popular_season = season_rentals.idxmax()
most_popular_season_name = season_mapping[most_popular_season]
print("Total rentals per season:")
print(season_rentals.rename(index=season_mapping))
print(f"\nThe season with the highest bike rentals is: {most_popular_season_name}")
```

## Q.17 What is the peak hour in which bike rents the most?

```python
hourly_rentals = df.groupby("hr")["cnt"].sum()
peak_hour = hourly_rentals.idxmax()
peak_hour_count = hourly_rentals.max()
print(f"Peak hour for bike rentals: {peak_hour}:00 with {peak_hour_count} rentals")
```

## Q.18 In which day of a week bikes rents out most?

```python
weekly_rentals = df.groupby("weekday")["cnt"].sum()
peak_day = weekly_rentals.idxmax()
peak_day_count = weekly_rentals.max()
print(f"Day with highest bike rentals: {peak_day} with {peak_day_count} rentals")
```

## Q.19 In which hour Casual users rents bikes the most?

```python
casual_rentals = df.groupby("hr")["casual"].sum()
peak_hour_casual = casual_rentals.idxmax()
peak_hour_casual_count = casual_rentals.max()

print(f"Peak hour for casual users: {peak_hour_casual}:00 with {peak_hour_casual_count} rentals")
```
